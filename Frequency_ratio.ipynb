```python
import numpy as np
import matplotlib.pyplot as plt

r = np.linspace(0,3,300) # frequency ratio
zeta = 0.25

a = 1-r**2
b = (2*zeta*r)

fr = 1/np.sqrt(a**2+b**2)

plt.figure(figsize=(7.5,5))
plt.plot(r,fr)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

r = np.linspace(0,4,300) # frequency ratio
zeta = [0.25,0.4,0.5,0.6]

plt.figure(figsize=(7.5,5))
for z in zeta:
    a = 1-r**2
    b = 2*z*r
    fr = 1/np.sqrt(a**2+b**2)
    plt.plot(r,fr,label=r'$\zeta={}$'.format(z))

plt.legend(loc='best')
plt.grid(which='major', color='black')
plt.minorticks_on()
plt.grid(which='minor', color='grey', linestyle = '--', alpha = 0.4)
plt.xlim(r[0], r[-1])
plt.show()
```
